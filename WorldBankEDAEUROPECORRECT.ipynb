```python
mask = df["Country Name"].isin(european_countries)
df = df[mask]
```

```python
for col in df.columns[5:]:
    df[col]=pd.to_numeric(df[col], errors=("coerce"))
```

```python
years = [
    "2000 [YR2000]", "2001 [YR2001]", "2002 [YR2002]", "2003 [YR2003]",
    "2004 [YR2004]", "2005 [YR2005]", "2006 [YR2006]", "2007 [YR2007]",
    "2008 [YR2008]", "2009 [YR2009]", "2010 [YR2010]", "2011 [YR2011]",
    "2012 [YR2012]", "2013 [YR2013]", "2014 [YR2014]", "2015 [YR2015]"
]
```

```python
df_long = df.melt(
    id_vars=["Country Name"],
    value_vars=years,
    var_name="Year",
    value_name="Life Expectancy"
)
```

```python
df_long["Year"] = df_long["Year"].str[:4].astype(int)
```

```python
df = df[df["Series Name"] == "Life expectancy at birth, total (years)"]
```

```python
year_cols = [col for col in df.columns if col.startswith("2")]
```

```python
df_long = df.melt(
    id_vars=["Country Name"],
    value_vars=year_cols,
    var_name="Year",
    value_name="Life Expectancy"
)
```

```python
df_long["Year"] = df_long["Year"].str[:4].astype(int)
```

```python
df_long.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))


sns.lineplot(data=df_long, x="Year", y="Life Expectancy", hue="Country Name", marker="o")

plt.title("Life Expectancy Over Time (2000-2015) - European Countries")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # legend'i saÄŸa taÅŸÄ±
plt.grid(True)
plt.tight_layout()
plt.show()
```
